## Cleaning

### Define

- Rename column with descriptive name for (ApplicationP, AboutC, RequiredQual, JobRequirment) and replace them with (AppilicationProcedure, AboutCompany, RequiredQualifications, JobRequirements)
- Select all records in startdate column that have 'As soon as possible', 'immediately', etc and replace them with ASAP

### Code

```python
df_clean = df.copy()
```

- Rename column with descriptive name for (ApplicationP, AboutC, RequiredQual, JobRequirment) and replace them with (AppilicationProcedure, AboutCompany, RequiredQualifications, JobRequirements)

```python
df_clean = df_clean.rename(columns={'RequiredQual': 'RequiredQualifications', 'ApplicationP': 'ApplicaitonProcedure', 'AboutC': 'AboutCompany', 'JobRequirment': 'JobRequirments'})
```

- Select all records in startdate column that have 'As soon as possible', 'immediately', etc and replace them with ASAP

```python
df_clean.StartDate.value_counts()
```

```python
# there are many similar phrases in the StartDate column and you have to list all of them down here we are just taking a sample of it
asap_list = ['Immediately', 'As soon as possible', 'Upon hiring', 'Immediate', 'Immediate employement', 'As soon as possible.']
```

```python
for phrase in asap_list:
    df_clean.StartDate.replace(phrase, 'ASAP', inplace=True)
```

### Test

```python
df_clean.info()
```

- So all the column headers are fixed

```python
df_clean.StartDate.value_counts()
```

- so we can see that the no of ASAP has increased significantly

```python
for phrase in asap_list:
    assert phrase not in df_clean.StartDate.values
```

- so no assertion error we are good, our cleaning code worked

## Analysis and Visualization

```python
asap_count = df_clean.StartDate.value_counts()['ASAP']
non_empty_count = df_clean.StartDate.count()
print(asap_count,non_empty_count)
```

```python
percentage = asap_count/non_empty_count
percentage
```

```python
%matplotlib inline
import numpy as np
labels = np.full(len(df_clean.StartDate.value_counts()), "", dtype=object)
labels[0] = 'ASAP'
df_clean.StartDate.value_counts().plot(kind='pie', labels=labels)
```
